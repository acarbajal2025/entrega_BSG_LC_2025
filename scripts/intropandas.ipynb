```python
# Seleccionar columnas 
df_aux = df[["StockCode", "Country"]]
df_aux.head()

```

```python
# seleccionas filas 
print(df.iloc[20:30])
```

```python
df_ventas_multiples = df[df["Quantity"] > 1]
df_ventas_multiples.sample(5)
```

```python
# Filtrado por Country
df_uk = df[ df['Country']=='United Kingdom']
df_uk.sample(5)
```

```python
df['Total price'] = df['UnitPrice'] * df['Quantity']
```

```python
df.head()
```

```python
Ventas_por_pais = df.groupby("Country")["Total price"].sum()
print(Ventas_por_pais)

```

```python
Ventas_por_producto = df.groupby("StockCode")["Total price"].sum()
print(Ventas_por_producto)
```
