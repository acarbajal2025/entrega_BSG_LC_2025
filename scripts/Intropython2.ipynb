```python
def mi_cuadrado(x):
    return x**2
    
```

```python
cuadrados = []
for i in range(10):
    cuadrados.append(mi_cuadrado(i))

print( cuadrados)
```

```python
# list comprehension 
cuadrados_2 = [mi_cuadrado(i) for i in range(10)]
print( cuadrados_2)
```

# Repaso de 
# range, tuples, lists, sets 

```python
my_range = range(5)  # lazy evaluation se hace solo cuando se necesita y no se guarda en memoria 
print('my_range :  ', my_range)
my_tuple = tuple(my_range)
print('my_tuple : ', my_tuple)
my_list  = list(my_tuple)
print('my_list : ', my_list)
my_set = set(range(5))
print('my_set is : ', my_set)
```

```python
# Range, usa "lazy evaluation" se hace solo cuando se necesita y no se guarda en memoria 
r = range(1, 30, 3) 
print(list(r))
print('start : ', r.start)  
print('stop : ', r.stop)   
print('step : ', r.step)   
```

```python
# Listas
my_list = list(my_tuple)
print('elemento con indice 4 : ', my_list[4])
# mutate
my_list[4] = 44
print('nuevo elemento con indice 4 : ', my_list[4])
# agregar elemento 
my_list.append(100)
print('my_list ahora es :', my_list)
# Extender la lista
my_list.extend( ['a','a','a','a','a', 'c'] )
print('my_list ahora es :', my_list)
# borrar elemento 
my_list.remove(3)
print('despues del remove my_list ahora es :', my_list)
# contar cuantas veces aparece un elemento
a_s = my_list.count('a')
print('cuantas veces esta a :', a_s)
```

```python
print(my_tuple)
# tuples son inmutables
print( my_tuple[3] )
my_tuple[3] = 33
```

```python
import timeit

# medir tiempo de ejecucion 
list_time = timeit.timeit(stmt="[1, 2, 3, 4, 5]", number=1000000) # number lo repite n veces para comparar mejor
tuple_time = timeit.timeit(stmt="(1, 2, 3, 4, 5)", number=1000000)

print('tiempo de ejecucion de lista es : ', list_time)
print("Para la tupla fue :", tuple_time) 
print('ratio es de :',  list_time / tuple_time)
```

```python
# En resumen: las listas son muy flexibles y se usan cuando quieres estarlas cambiando, su uso es mas lento
# Las tuplas se usan cuando quieres un objeto inmutable y mas eficiente
# Los rangos se usan para generarse en el momento y no se guardan en memoria, usan "lazy evaluation"
```
